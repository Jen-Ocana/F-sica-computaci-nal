```

import pandas as pd
import numpy as np
url= 'https://raw.githubusercontent.com/Jen-Ocana/Fisica-Computacional-1/master/Actividad%203/8352.txt'
my_cols = ['Fecha','Precip','Evap','Tmax','Tmin']

df_orig= pd.read_csv(url, names=my_cols, encoding='cp1251', sep='\s+', header=None, skiprows=19, skipfooter=1, engine='python')

################ARCHIVO ORIGINAL########################

#      al 20 sale todo y dos reglones de datos
#      reglon 18 o 19 contando el cero esta el primer dato
#      Reglon 17 vacio  18 contando el cero
#      primer numero/numero considerando el 0
#      Reglon 16/17 los nombres de las columnas no tienen la columan de fecha que
#      es la primera

##############INSTRUCCIONES DE LA ACTIVIDAD ##############
#Imprime el encabezado y final del dataframe: 
print("------Encabezado del dataframe---------- \n",df_orig.head(10))
print("")
print("------Final del dataframe---------- \n",df_orig.tail(10))
print("") 
#¿Qué dimensiones tiene tu dataframe?:
print("----------Dimensiones del data frame------\n",df_orig.shape,end="\n")
print("")
#¿Cómo es el contenido de tu dataframe?: \n
print("----------Informacion del data frame------\n")
print(df_orig.info(),end="\n")
print("")
#Los datos originales incluyen la cadena de caracteres 'Nulo', indicando que no 
#hubo datos para esa variable, ese día. Tenemos que reemplazar la palabra nulo 
#con la función: df.replace()
df_work0 = df_orig.copy()
df_work0.head()

str_Nulo = 'Nulo'
df_work1 = df_work0.replace(to_replace=str_Nulo, value='', regex=True)
df_work1.head()

print("--------Dataframe:Work1 info-------\n")
print(df_work1.info())
print("")

print("------Encabezado del dataframe nuevo--------- \n",df_work1.head())
print("")
#Después habrá que convertir a número flotante o numérico los datos de
# Precipitación, Evaporación, Temperatura Máxima y Temperatura mínima 
#utilizando la función: df.to_numeric()
#_________________________________________
## Deseamos convertir los valores de las variables ambientales a núm. flotantes 
cols_list = ['Precip', 'Evap', 'Tmax', 'Tmin']
for cols in cols_list:
  df_work1[cols] = pd.to_numeric(df_work1[cols], errors='coerce')

#Contrasta ahora la información de tu DataFrame con la función: df.info() 

print("------Informacion de la actualizacion---------\n")
print(df_work1.info()) 
##Aqui podemos ver arriba todavia que el tipo ya no son 5 objetos
print("")

#Se puede contabilizar el número de datos faltantes en esas variables mediante la función: df.isnull().sum()
# De la instrucción anterior, vemos que hay valores faltantes. Sumamos los valores nulos. 
print("------Sumando los valores nulos------\n")
print(df_work1.isnull().sum())
print("")

#Imprime de nuevo el encabezado y final de tu dataframe, con las funciones df.head() y df.tail()
print("------Encabezado del work1---------- \n",df_work1.head()) 
print("")
print("------Final del work1---------- \n",df_work1.tail())
print("")


#Realiza una estadística básica de las variables numéricas de tu dataframe
# usando la función: df.describe() y haz una interpretación de los resultados, 
#para ver si tienen sentido físico (por ej. valores negativos de precipitación, 
#valores extremos fuera de lo normal, etc)
print("------ Descripcion ---------- \n",df_work1.describe().round(2))
print("")
#Aqui todo parece tener sentido


##########---Hacemos work2 para modificar la fecha----######
###Escribimos las instrucciones del paso 4

print("###########PASO 4 MODIFICANDO LA FECHA############")
print("")

#Comienza haciendo una copia del dataframe del paso anterior
#, por si se requiere recuperar el nuevo dataframe,
# utilizando df:new=df.copy()

df_work2=df_work1.copy()
print("Vemos la columan de fecha solamente \n")
print(df_work2['Fecha'].head())
print("")

#Utiliza la función de Pandas pd.to_datetime(), 
#para convertir el objeto Fecha a formato de fecha 
#que comprende Python.  Lee el manual de esa función e 
#intenta la conversión. 
#df_work2['Fecha']=pd.to_datetime(df_work2['fecha'],dayfirst=True).copy()
df_work2['Fecha']=pd.to_datetime(df_work2['Fecha'], dayfirst=True).copy()
print("------ Encabezado correjida la fecha ---------- \n")
print("")
print(df_work2.head())
print("------- Encabezado del work2 ----------\n")
print(df_work1.head())
##Para comparar
print("")

#Después utiliza la función df.dtypes para verificar que todas 
#las variables son del tipo deseado. 
print("------ Checando tipos de la fecha ---------- \n")
print(df_work2.dtypes)
print("")

#Enseguida con la ayuda de las funciones de 
#Pandas df['Fecha'].dt.year y df['Fecha'].dt.month, crea 
#dos columnas nuevas adicionales df['Año'] y df['Mes'].
df_work2['Año'] = df_work2['Fecha'].dt.year
df_work2['Mes'] = df_work2['Fecha'].dt.month

#Utilizando la función print aplicado a las funciones 
#df.head() y df.tail(), verifica que el dataframe tiene 
#la forma deseada.
print("------------Encabezado con las columnas de año---------\n")
print(df_work2.head())
print("")
print("------------Final con las columnas de año-------------\n")
print(df_work2.tail())
print("")

#Y complementa con la función df.dtypes, para verificar que todas
# las variables son del tipo adecuado. 
print("----Comprobamos que todas las variables sean del tipo deseado-----")
print("")
print(df_work2.info())
print("")
print("***********Terminamos*************")

#Hasta aquí llegaremos esta semana. 



```
